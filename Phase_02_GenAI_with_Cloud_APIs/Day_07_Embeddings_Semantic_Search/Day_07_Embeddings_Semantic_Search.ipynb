```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer

import chromadb
```

```python
documents=[
    "ETL is used to clean & transform data",
    "SQL is used to query databases",
    "cars & trucks are popular  automobiles"
    "Machine learning train models on data"
]
query_keyword="vehicle"
print("Keyword search for :",query_keyword)

for i,doc in enumerate(documents):
    if query_keyword.lower() in doc.lower():
        print(f"FOUND [doc{i}]:{doc}")
    else:
        print(f"MISSED [doc{i}]:{doc}")
```

```python
failure_examples=[
    {"query":"I feel sick", "misses":"I am unwell,patient has fever"},
    {"query":"How to cook rice","misses":"Steps to prepare rice"},
    {"query":"Vehicle speed","misses":"car acceleration,automobile velocity"},
    {"query":"ML Model accuracy","misses":"Classification performance"},
]
print("KEYWORD SEARCH FAILURE CASES")
for ex in failure_examples:
    print(f"Query:{ex['query']}")
    print("Misses",ex['misses'])
```

```python
print("Loading embedding model...(may take 1-2 mins on first run)")

model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentence="ETL is used to clean & transform data"
embedding=model.encode(sentence)
print(embedding.shape)
print(embedding)
print("Min value",embedding.min())
print("Max value",embedding.max())
```

```python
sentences=[
    "ETL is used to clean & transform data",
    "SQL is used to query databases",
    "cars & trucks are popular  automobiles"
    "Machine learning train models on data",
]
embeddings=model.encode(sentences)
print("Number of sentences:",len(sentences))
```

```python
print("Each row is one sentence embedding")
for i,sent in enumerate(sentences):
    print(f"sentence{i}:shape={embeddings[i].shape}")
```

```python
def cosine_similarity(vec_a,vec_b):
    dot_product=np.dot(vec_a,vec_b)
    norm_a=np.linalg.norm(vec_a)
    norm_b=np.linalg.norm(vec_b)
    similarity=dot_product/(norm_a*norm_b)
    return similarity
sim_01=cosine_similarity(embeddings[0],embeddings[1])
sim_02=cosine_similarity(embeddings[0],embeddings[2])
sim_12=cosine_similarity(embeddings[1],embeddings[2])
print()
print(f"Similarity (0 vs 1):{sim_01}")
print(f"Similarity (0 vs 2):{sim_02}")
print(f"Similarity (1 vs 2):{sim_12}")
```

```python
your_sentences=[
    "Machine learning trains models on labeled data",
    "AI alogorithms learn patterns for examples",
    "I enjoy eating pizza for lunch"
]
your_embeddings=model.encode(your_sentences)
sim_your_01=cosine_similarity(your_embeddings[0],embeddings[0])
sim_your_02=cosine_similarity(your_embeddings[0],embeddings[1])
sim_your_03=cosine_similarity(your_embeddings[0],embeddings[2])
print("YOUR EXPERIMENT RESULTS")
print(f"Sentence A: {your_sentences[0]}")
print(f"Sentence B: {sentences[0]}")
print(f"Similarity: {sim_your_01}")
```

ChromaDB
It is a vector DB
Instead of using traditional rows and columns, ChromaDB translates unstructured data (text, images, audio) into mathematical lists of numbers called embeddings. When you query the database, it performs a similarity search to find the stored vectors that most closely match your query.



*   Collection: Like a table in SQL-a named group of documents
*   Document: THe text of your note
* A unique identifier for each document
* Metadata: Extra info about the document
* Distance: How far each vectors are



```python
chroma_client=chromadb.Client()
collection=chroma_client.get_or_create_collection("demo_notes")

```

```python
sample_docs=[
    "ETL stands for Extract Transform Load- the core data engineering process",
    "SQL SELECT statements retreive data from database tables",
    "Machine learning models learn paterns from training data",
    "Python pandas library is used for data manipulation and cleaning",
    "Neural networks are inspired by how human brain works",
]
sample_ids=["doc001","doc002","doc003","doc004","doc005"]
sample_metadata=[
    {
        "subject":"Data engineering","topic":"ETL"
    },
    {
        "subject":"Data engineering","topic":"SQL"
    },
    {
        "subject":"Machine learning","topic":"ML basics"
    },
    {
        "subject":"Python","topic":"Pandas"
    },
    {
        "subject":"Machine learning","topic":"Neural networks"
    }
]

collection.add(
    documents=sample_docs,
    ids=sample_ids,
    metadatas=sample_metadata
)
```

```python

query="How do I clean & prepare data?"
results=collection.query(
    query_texts=query,
    n_results=3
)
print("Query:",query)
print()

matched_docs=results['documents'][0]
matched_ids=results['ids'][0]
matched_distances=results['distances'][0]
matched_metadatas=results['metadatas'][0]

for rank,(doc,doc_id,dist,meta) in enumerate(zip(matched_docs,matched_ids,matched_distances,matched_metadatas)):
    print(f"Rank {rank+1}")
    print("Document:",doc)
    print("Document ID:",doc_id)
    print("Distance:",dist)
    print("Metadata:",meta)
```

```python
filtered_query="How do computers learn from examples?"
filtered_results=collection.query(
    query_texts=filtered_query,
    n_results=3,
    where={"subject":"Machine learning"}
)
print("Filtered Query:",filtered_query)

for rank,(doc,dist,meta) in enumerate(zip(
    filtered_results['documents'][0],
    filtered_results['distances'][0],
    filtered_results['metadatas'][0]
),start=1):
    print(f"Rank {rank+1}")
    print("Document:",doc)
    print("Distance:",dist)
    print()
```

```python
print("DISTANCE TO SIMILARITY CONVERSION")

print(f"{'Distance':<15}{'Similarity':<15}{'Interpretation:':<20}")
distances=[0.20,0.40,0.65,0.90]
interpretation=["Near identical","Very similar","Related","Somewhat related",
                "Not related"]

for dist,interp in zip(distances,interpretation):
    similarity=1-dist
    print(f"{dist:<15.2f}{similarity:<15.2f}{interp:<20}")
```
