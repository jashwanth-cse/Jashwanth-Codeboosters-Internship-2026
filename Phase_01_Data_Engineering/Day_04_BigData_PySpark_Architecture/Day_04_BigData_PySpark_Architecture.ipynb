```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("burnoutminer/heights-and-weights-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# The 'path' variable from the previous cell holds the directory path.
# We need to find the actual CSV file within this directory.

csv_file_path = None
# List the contents of the directory to find the CSV file
for filename in os.listdir(path):
    if filename.endswith(".csv"):
        # Construct the full path to the CSV file
        csv_file_path = os.path.join(path, filename)
        break

if csv_file_path:
    df = pd.read_csv(csv_file_path)
    # Display the first 5 rows to verify
    display(df.head())
else:
    print(f"Error: No CSV file found in the directory '{path}'")
    raise FileNotFoundError(f"No CSV file found in the directory '{path}'")
```

```python
x=df[["Height(Inches)"]]
y=df["Weight(Pounds)"]
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
model.predict(x_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
